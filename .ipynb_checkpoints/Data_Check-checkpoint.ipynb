Unique SEQN entries = 15560 but total rows = 32962. 32962 is consistent with what is displayed for cumulative responses for rx information on CDC website. 

In reading further: "Respondent sequence number (SEQN) in the prescription medication data is not a unique identifier. Participants who reported the use of multiple prescription drugs will have multiple records, that is, a record for each drug. Analysts should convert a drug-level data to a person-level data, that is, a record for each person, before merging it with NHANES demographic and other data files by using SEQN."

So all merged data will have multiple entries due to the SEQN not being a unique identifier for rx data. 

Possible Solutions: 
1) eliminate duplicates, since the columns that we are interested in should have the same values (# of meds taken will be consistent, # days taken will not)
2) unmerge the data - take out meds and sleep, analyze separately given issues with duplicates and encoding

```python
#based on reasoning above, removed dups from demographic data to check counts
demographics_check = Check_dataset.drop(columns=['Taken Rx Past Month', 'Number Days Taken Rx',
       'Number of Rx Meds', 'Vigorous Work Activity', 'Number of Days Vigorous Work',
       'Minutes of Vigorous Work Daily', 'Moderate Work Activity',
       'Number of Days Moderate Work', 'Minutes of Moderate Work Daily',
       'Walk or Bike Travel', 'Number of Days Walk or Bike Weekly',
       'Daily Minutes Walking or Biking', 'Vigorous Recreational Activity',
       'Number of Days Vigorous Recreation',
       'Minutes of Vigorous Recreation Daily',
       'Moderate Recreational Activity', 'Number of Days Moderate Recreation',
       'Minutes of Moderate Recreation Daily',
       'Minutes of Sedentary Activity Daily', 'Bedtime Week/Workdays',
       'Waketime Week/Workdays', 'Sleep Hours Week/Workdays',
       'Bedtime Weekends', 'Waketime Weekends', 'Sleep Hours Weekends',
       'Snoring Frequency Past Year',
       'Frequency Snort/Stop Breathing Past Year',
       'Told Doctor About Sleep Issues', 'Feeling Tired During Day Frequency'])
   
demographics_check
```

```python
demo_remove_dups = demographics_check.drop_duplicates(subset=['Respondent ID'], keep='last')
demo_remove_dups
```

```python
demo_remove_dups.count()
```

Demographics data pass the check, as unique SEQN identifiers for this dataset = 15560 which is the reported cumulative number of responses for demographics data. This makes sense as gender was chosen for all respondents.
