Removed duplicate entries caused by ascertainment of data in one of the modules (Prescription Medications)

```python
# Dependencies
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
import pyreadstat
import xport
```

```python
# Data_update import 
Dataset_update = pd.read_csv("Datasets/Dataset_update")

# Display dataframe
Dataset_update
```

See "Data_Check" for the analysis performed to check that the data imported correctly. During this analysis, we discovered that SEQN (Respondent ID) cannot be used as a unique identifier for the Prescription Meds dataset because a single respondent may take multiple medications, resulting in multiple rows per respondent (see Respondent ID 124821.0 for example). Because the specific data point we want to analyze with regards to Prescription Meds will remain consistent across multiple entries per respondent, removing duplicate Responent IDs seemed to be the most straightforward approach. 

```python
#Based on above analysis description, removed duplicates within the Respondent ID column
Dataset_update_remove_dups = Dataset_update.drop_duplicates(subset=['Respondent ID'], keep='last')
Dataset_update_remove_dups
```

```python
#checked the counts of each column against published data counts on CDC website
Dataset_update_remove_dups.count()
```

After removing duplicates, the counts per column match what is published on the CDC website with the exception of "Number of Rx Meds" which is expected given the multiple entries per respondent issue.

```python
#wrote new data set to csv
Dataset_update_remove_dups.to_csv("Datasets/Dataset_update_remove_dups",
                  encoding="utf-8", index=False, header=True)
```
